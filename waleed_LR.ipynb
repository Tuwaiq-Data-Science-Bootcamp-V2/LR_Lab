## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

## Q9: Train the Model.

```python
lm.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lm.intercept_)
```

```python
Coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
Coeff_df
```

```python
preductions = lm.predict(X_test)
preductions
```

```python
real_value = np.array(y_test)
real_value
```

```python
plt.scatter(real_value,preductions)
```

## Residual Histogram

```python
sns.distplot((y_test-preductions))
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, preductions))
print('MSE:', metrics.mean_squared_error(y_test, preductions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, preductions)))
```

```python
print('R^2 : ',metrics.r2_score(y_test,preductions))
```
